# Data Visulaization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
num_features = len(data.columns) - 1

rows = num_features
columns = num_features
data.drop('Heart Disease',axis=1).hist(bins=15,figsize=(25,rows*3),layout=(rows,columns))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Heart Disease',data=data)
plt.title('Frequency of Heart Disease Categories')
plt.show()
```

```python
data['Heart Disease']=data['Heart Disease'].apply(lambda x:1 if x=="Presence" else 0)
correlation_matrix=data.corr()

plt.figure(figsize=(12,9))
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',fmt='.2f',linewidth=5)
plt.title('Correlation Matrix of Heart Disease DataSet')
plt.show()
```
