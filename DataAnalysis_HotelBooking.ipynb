# Data Analysis and Visualizations

```python
cancelled_perc=df['is_canceled'].value_counts(normalize=True)
print(cancelled_perc)

plt.figure(figsize=(5,4))
plt.title('Reservation status count')
plt.bar(['Not canceled','Canceled'],df['is_canceled'].value_counts(),edgecolor='k',width=0.7)
plt.show()
```

```python
plt.figure(figsize=(8,4))
ax1=sns.countplot(x='hotel',hue='is_canceled',data=df,palette='Blues')
plt.title('Reservation status in different hotels',size = 20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.show()
```

```python
resort_hotel=df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel=df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in City and Resort Hotel',fontsize = 20)
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
ax1=sns.countplot(x='month',hue='is_canceled',data=df,palette='bright')
plt.title('Reservation status per month',size =20)
plt.xlabel('month')
plt.ylabel('number of reservations')
plt.legend(['not canceled','canceled'])
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.title("ADR per month",fontsize=30)
cancelled_bookings = df[df['is_canceled'] == 1]
monthly_adr_sum = cancelled_bookings.groupby('month')['adr'].sum().reset_index()
sns.barplot(x='month', y='adr', data=monthly_adr_sum)
plt.xlabel("Month")
plt.ylabel("ADR")
plt.xticks(rotation=45) 
plt.show()
```

```python
cancelled_bookings = df[df['is_canceled'] == 1]
top_10_country=cancelled_bookings['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title('Top 10 countries with reservation cancelled')
plt.pie(top_10_country,autopct='%.2f',labels=top_10_country.index)
plt.show()
```

```python
cancelled_df_adr=cancelled_bookings.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date',inplace=True)
notcancelled_bookings=df[df['is_canceled']== 0]
notcancelled_df_adr=notcancelled_bookings.groupby('reservation_status_date')[['adr']].mean()
notcancelled_df_adr.reset_index(inplace=True)
notcancelled_df_adr.sort_values('reservation_status_date',inplace=True)
# Plot settings
plt.figure(figsize=(15, 8))
plt.title("Average Daily Rate", fontsize=20)
plt.xlabel("Reservation Status Date", fontsize=14)
plt.ylabel("Mean ADR", fontsize=14)

# Plot canceled bookings
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='Canceled Bookings')

# Plot not canceled bookings
plt.plot(notcancelled_df_adr['reservation_status_date'], notcancelled_df_adr['adr'], label='Not Canceled Bookings')

# Add legend
plt.legend()

# Show plot
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter canceled bookings for the period between 2016 and September 2017
cancelled_df_adr = cancelled_bookings[(cancelled_bookings['reservation_status_date'] >= '2016-01-01') & 
                                      (cancelled_bookings['reservation_status_date'] < '2017-09-30')]
cancelled_df_adr = cancelled_df_adr.groupby('reservation_status_date')[['adr']].mean().reset_index()

# Filter not canceled bookings for the period between 2016 and September 2017
notcancelled_df_adr = notcancelled_bookings[(notcancelled_bookings['reservation_status_date'] >= '2016-01-01') & 
                                            (notcancelled_bookings['reservation_status_date'] < '2017-09-30')]
notcancelled_df_adr = notcancelled_df_adr.groupby('reservation_status_date')[['adr']].mean().reset_index()

# Plot settings
plt.figure(figsize=(15, 8))
plt.title("Average Daily Rate", fontsize=20)
plt.xlabel("Reservation Status Date", fontsize=14)
plt.ylabel("Mean ADR", fontsize=14)

# Plot canceled bookings
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='Canceled Bookings')

# Plot not canceled bookings
plt.plot(notcancelled_df_adr['reservation_status_date'], notcancelled_df_adr['adr'], label='Not Canceled Bookings')

# Add legend
plt.legend()

# Show plot
plt.show()
```
